```python
AREA_PUERTA = 2.0
AREA_VENTANA = 1.0

Largo = float(input("Ingrese el largo de la habitación:  "))
Ancho = float(input("Ingrese el ancho de la habitación:  "))
Alto = float(input("Ingrese el alto de la habitación:  "))

Area_Paredes = 2 * (Largo * Alto) + 2 * (Ancho * Alto)
Area_Techo = Largo * Ancho
superficie_total = Area_Paredes + Area_Techo

puertas = int(input("Ingrese el número de puertas:  "))
ventanas = int(input("Ingrese el número de ventanas:  "))
areas_huecos = puertas * AREA_PUERTA + ventanas * AREA_VENTANA
superficie_total -= areas_huecos

print("La superficie total a pintar es:", superficie_total, "metros cuadrados.")

costo_metro_cuadrado = float(input("Ingrese el costo por metro cuadrado de pintura:  "))
costo_total = superficie_total * costo_metro_cuadrado

print("El costo total de pintura es:", costo_total, "euros.")

efectivo = float(input("Ingrese el efectivo proporcionado:  "))
cambio = efectivo - costo_total

if cambio < 0:
    print("No hay suficiente efectivo para cubrir el costo.")
else:
  print("El cambio es:", cambio, "euros.")

  cent_restantes = cambio
  monedas_25 = int(cent_restantes // 25)
  cent_restantes -= monedas_25
  monedas_10 = int(cent_restantes // 10)
  cent_restantes -= monedas_10
  monedas_5 = int(cent_restantes // 5)
  cent_restantes -= monedas_5
  monedas_1 = int(cent_restantes)

  print("Cambio en monedas de 25 centavos:", monedas_25)
  print("Cambio en monedas de 10 centavos:", monedas_10)
  print("Cambio en monedas de 5 centavos:", monedas_5)
  print("Cambio en monedas de 1 centavo:", monedas_1)
```
